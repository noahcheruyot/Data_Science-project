```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load your dataset
data = pd.read_excel('/content/Accident_Severely_prediction.csv.xlsx')
data

# Define the independent variables (features) and the dependent variable (target)
X = data["Weather", "Time_of_Day", "Road_Conditions", "Speed_Limite", "Num_Vehicles", "Accident_Severity"]
y = data['Severity']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=52)

# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)

# Example of predicting accident severity for new data
new_data = pd.DataFrame({'Weather': ['Clear'],
                         'Road Conditions': ['Dry'],
                         'Driver Age': [30],
                         'Time of Day': ['Day'],
                         'Speed Limit': [45],
                         'Alcohol Involvement': [0],
                         'Vehicles Involved': [2]})
predicted_severity = model.predict(new_data)
print("Predicted Severity for New Data:", predicted_severity)
```
